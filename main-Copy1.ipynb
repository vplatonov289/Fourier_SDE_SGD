### Initialising libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class SDE ():
    
    ### Check the correct passing of the arguments
    
    def __init__(self, x_0 = 0, b = 1, sigma = 1, T = 1, n_discr = 100):
        self.x_0 = x_0
        self.b = b
        self.sigma = sigma
        self.T = T
        self.n_discr = n_discr
        self.dt = self.T / self.n_discr
        self.x = self.get_path()
        
        
    def get_path(self):
        x = np.zeros(n_discr)
        x[0] = self.x_0 
        dW_t = np.random.normal(size = n_discr, scale = np.sqrt(self.dt))
                
        for i in range(1,n_discr):
            x[i] = x[i - 1] + self.b[i - 1] * x[i - 1] * self.dt + self.sigma[i - 1] * x[i - 1] * dW_t[i - 1]
            
        return x
            
    def plot(self):
        
        t = np.linspace(0, self.T, self.n_discr)
        
        fig, ax = plt.subplots(1,1,figsize=(15, 10), tight_layout=True)

        ax.set_title(r"Dynamics of an SDE", fontsize = 15)
        ax.set_xlabel(r'$t$',fontsize=15)
        ax.set_ylabel(r'$X_t$',fontsize=15)
        ax.tick_params(axis='both', which='major', labelsize = 20)
        ax.tick_params(axis='both', which='minor', labelsize = 20)
        ax.plot(t, self.x)
        
        plt.show()
```

```python
n_discr = 10000
T = 1
x_0 = 1

b = np.ones(n_discr) - 1.5
sigma = np.ones(n_discr)

X = SDE(x_0, b, sigma, T, n_discr)
X.plot()
```
