<font size="4"> $$Y=X+(-Y^3)h \theta  $$
$$ h\cdot Y^3+Y-X=0,\qquad a=\theta h,b=0,c=1,d=-x$$

```python

def make_brownian_motion(bm,h):  # Update BM, push it into h future.
    hh=np.sqrt(h)
    for i in range(len(bm)):
        rrr=hh*np.random.normal(0,1)
        bm[i]+=rrr 
    return(0)

def taming_method(x,bm,h):     ## x[i][0] is BM information, x[i][1] is particle's information  
    ex=0
    n=len(bm)
    sig=1.5
    c=0.5
    T=1
    M=(T/h)
    alpha=0.5
    mma=M**(-alpha)
    sig22=0.5*sig*sig
    for i in range(len(bm)):
        ex+=x[i][1]
    cex=c*ex/n
    for i in range(0,n):
        bbb=(sig22*x[i][1]-(x[i][1]**3)+cex)
        bbb=bbb/(1+mma*abs(bbb))
        dWt=bm[i]-x[i][0]
        x[i][0]=bm[i]                  # Update BM 
        x[i][1]+=bbb*h + sig*x[i][1]*dWt

def split_step_method(x,bm,h,theta): # This is the split step method but with theta, theta=1 is in the article, but it can be 0.5-1.
    ey=0
    sig=1.5
    c=0.5
    y=[]
    n=len(bm)
    twoh=2*h*theta
    threeh=3*h*theta
    sig22=0.5*sig*sig
    ## a=h, b=0, c=1 ,d=-x  This part is the implcit solution of the cubic function. the y star in the paper.
    secondtermofdelta=(1/threeh)**3
    for i in range(0,n):
        delta_sqrt=np.sqrt(  (x[i][1]/twoh)**2 +  secondtermofdelta  )
        y.append(0)
        part1=(-(-x[i][1]/twoh)+ delta_sqrt)
        part2=(-(-x[i][1]/twoh)- delta_sqrt)
        if ( part1>=0):
            y[i]+=part1**(1/3)
        else:
            y[i]+=-(-part1)**(1/3)
        if ( part2>=0):
            y[i]+=part2**(1/3)
        else:
            y[i]+=-(-part2)**(1/3)                
            
        ey+=y[i]
    cey=c*ey/n
    for i in range(0,n):
        bbb=(sig22*y[i]+cey)
        dWt=bm[i]-x[i][0]
        x[i][0]=bm[i]                  # Update BM 
        x[i][1]+=-(y[i]**3)*h+bbb*h + sig*y[i]*dWt        

        
def adaptive_step_method(x,bm,h): # Adaptive method. This is not exactly right, the BM shall change as time, but it doesn'r change the result.
    ex=0
    hh=np.sqrt(h)
    sig=1.5
    c=0.5
    n=len(bm)
    N=len(bm)
    rou=int(1/h)
    sig22=0.5*sig*sig
    for i in range(0,N):
        ex+=x[i][1]
    cex=c*ex/n     
    for i in range(N):
        done=True
        #done=False
        countt=0
        dWt=bm[i]-x[i][0]
        x[i][0]=bm[i] 
        difusion_increment=dWt*x[i][1]*sig
        #x[i]+=((sig*sig*0.5*x[i])-(x[i]**3)+c*ex)*h + sig*x[i]*bm[i][count]  
        while done:
            #print('xxx')
            dh=h*min(1,x[i][1]**(-2))
            #dh=h
            if dh < h-countt:
                bbb=sig22*x[i][1]-(x[i][1]**3)+cex
                x[i][1]+=bbb*dh
                countt+=dh
            else:
                dh=h-countt
                #print('xxx',dh)
                bbb=(sig22*x[i][1]-(x[i][1]**3)+cex)
                x[i][1]+=bbb*dh
                countt+=dh    
                done=False           
        x[i][1]+=difusion_increment

```

```python
#Save/Load the particle information to excel.

workbook = xlwt.Workbook(encoding = 'utf-16')
worksheet = workbook.add_sheet('My Worksheet')
#data= xlrd.open_workbook("x.xlsx")
#table = data.sheets()[0]
########################################



def savefile_single(x,count):
    for i in range(len(x)):
        worksheet.write(i,count,x[i])
    count+=1
    return(count)

def savefile_double(x,count):
    for i in range(len(x)):
        worksheet.write(i,count,x[i][0])
    count+=1
    for i in range(len(x)):
        worksheet.write(i,count,x[i][1])
    count+=1
    return(count)

def filetolist(x,count):
    data= xlrd.open_workbook("Simulation20210419_1.xls")
    table = data.sheets()[0]
    for i in range(len(x)):
        x[i][0]=table.cell_value(i,count)
        x[i][1]=table.cell_value(i,count+1)
    return(count+2)
```

```python
#Error calculation
def calculation_strong_error(exact,approx):
    n=len(exact)
    strong_error=0
    for i in range(n):
        strong_error+=(exact[i][1]-approx[i][1])**2
    strong_error=np.sqrt(strong_error/n)
    return(strong_error)

def calculation_weak_error(exact,approx):
    n=len(exact)
    weak_error=0
    for i in range(n):
        weak_error+=(exact[i][1]-approx[i][1])
    weak_error=abs(weak_error/n)
    return(weak_error)
```

```python
## This is the simulation
N=1000
T=1
theta1=0.6
theta2=1
bm=[0 for i in range(N)]
particle_taming_method=[[[0,1] for i in range(N)] for j in range(10)]
particle_split_step_method1=[[[0,1] for i in range(N)] for j in range(10)]
particle_split_step_method2=[[[0,1] for i in range(N)] for j in range(10)]
particle_adaptive_timestep_method=[[[0,1] for i in range(N)] for j in range(10)]
particle_splitting_method=[[[0,1] for i in range(N)] for j in range(10)]


particle_taming_exact=[[0,1] for i in range(N)] 
particle_split_step_exact1=[[0,1] for i in range(N)] 
particle_split_step_exact2=[[0,1] for i in range(N)] 
particle_adaptive_exact=[[0,1] for i in range(N)]
particle_splitting_exact=[[0,1] for i in range(N)]


time_taming=[0 for i in range(11)]
time_split_step1=[0 for i in range(11)]
time_split_step2=[0 for i in range(11)]
time_adaptive=[0 for i in range(11)]
time_splitting=[0 for i in range(11)]

timestep=[0.0001,0.0002,0.0005,0.001,0.002,0.005,0.01,0.02,0.05,0.1]
moder=[0.0001,0.0002,0.0005,0.001,0.002,0.005,0.01,0.02,0.05,0.1]
# Exact solution is calculate under h=10^{-5}
for i in range(10):
    moder[i]=int(moder[i]*(10**5))        # this sequence is for the count of the timestep, w.r.t the smallest timestep. 
    
for i in range(1,1*(10**5)+1):
    #print(i)
    if (i%1000==0): print("OK",i)
    make_brownian_motion(bm,0.00001)  #make B.M.
    t1=time.time()
    taming_method(particle_taming_exact,bm,0.00001)
    t2=time.time()
    split_step_method(particle_split_step_exact1,bm,0.00001,theta1)
    t3=time.time()
    adaptive_step_method(particle_adaptive_exact,bm,0.00001)
    t4=time.time()
    splitting_explci_tam_method(particle_splitting_exact,bm,0.00001)
    t5=time.time()
    split_step_method(particle_split_step_exact2,bm,0.00001,theta2)
    t6=time.time()
    time_taming[0]+=t2-t1
    time_split_step1[0]+=t3-t2
    time_adaptive[0]+=t4-t3
    time_splitting[0]+=t5-t4
    time_split_step2[0]+=t6-t5
    
    for j in range(10):    # Same BM but unde different timestep scale.
        if (i%moder[j]==0):    # To see if we are at the right timestep and undate the correspond process.
            t1=time.time()
            taming_method(particle_taming_method[j],bm,timestep[j])
            t2=time.time()
            split_step_method(particle_split_step_method1[j],bm,timestep[j],theta1)
            t3=time.time()
            adaptive_step_method(particle_adaptive_timestep_method[j],bm,timestep[j])
            t4=time.time()
            splitting_explci_tam_method(particle_splitting_method[j],bm,timestep[j])
            t5=time.time()
            split_step_method(particle_split_step_method2[j],bm,timestep[j],theta2)
            t6=time.time()
            
            time_taming[j+1]+=t2-t1
            time_split_step1[j+1]+=t3-t2
            time_adaptive[j+1]+=t4-t3
            time_splitting[j+1]+=t5-t4
            time_split_step2[j+1]+=t6-t5
```
